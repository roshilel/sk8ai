```python
import os
import subprocess
```

```python
PROJ_DIR = '/content/drive/My Drive/sk8ai/'
FULL_VIDS_DIR = os.path.join(PROJ_DIR, "full_vids")
KICKFLIP_CLIPS_DIR = os.path.join(PROJ_DIR, "clips/kickflip/")


kickflip_txt = os.path.join(PROJ_DIR, 'txt_data/kickflip.txt')

with open(kickflip_txt, 'r') as f:
  for i, line in enumerate(f):
    url, label, start_time, end_time = line.split(',')
    vid_id = url.split('v=')[1].split('&')[0] #get vid id, remove & params
    full_vid_path = os.path.join(FULL_VIDS_DIR, f'{vid_id}.mp4')

    if not os.path.exists(full_vid_path):
      yt_dlp_command = [
          'yt-dlp',
          '-f', 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]/best',
          '-o', full_vid_path,
          url
      ]
      subprocess.run(yt_dlp_command, check=True, capture_output=True, text=True)
      print(f'> Downloaded {full_vid_path}')

    else:
      print(f'> {full_vid_path} already exists. Skipping download.')

    #00:00:00.000
    #012345678901
    #          11
    format_time = lambda time : time[:2]+'h'+time[3:5]+'m'+time[6:8]+'s'+time[9:]+'ms'
    output_clip_path = os.path.join(KICKFLIP_CLIPS_DIR, f"{label}_{vid_id}_{format_time(start_time)}_{format_time(end_time)}.mp4")

    ffmpeg_command = [
        'ffmpeg',
        '-i', full_vid_path,
        '-ss', start_time,
        '-to', end_time,
        '-c', 'copy',
        '-y',
        output_clip_path
    ]
    subprocess.run(ffmpeg_command, check=True, capture_output=True, text=True)

    print(vid_id)
    break
```
